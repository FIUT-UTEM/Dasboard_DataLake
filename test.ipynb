```python
import pandas as pd
from sqlalchemy import create_engine

# Crear conexión usando SQLAlchemy
engine = create_engine('mariadb+mariadbconnector://testfiut:utem1234@localhost/mysql')
```

```python
pd.read_csv('data/porcentajes avances.csv', sep=';')
```

```python
query_fecha = "SELECT MAX(fecha_actualizacion) as ultima_fecha FROM fiut.avance_indicadores"
df_fecha = pd.read_sql(query_fecha, engine)
ultima_fecha = df_fecha['ultima_fecha'].iloc[0]

# Obtener los datos con la fecha más reciente
query = f"""
    SELECT 
        ai.id_indicador 'ID', 
        ai.dimension AS 'Dimension', 
        ai.estado AS 'Estado', 
        ai.origen 
    FROM  
        fiut.avance_indicadores ai
    WHERE 
        ai.fecha_actualizacion = (select max(aid.fecha_actualizacion) from fiut.avance_indicadores as aid)
"""
df_avances = pd.read_sql(query, engine)
```
